In the below I am using decalring the class and objects and calling them later for the use:: The __init__ will be like executed by default if we call this particular class.
This __init__ can be used to initialize any other common steps like having the database created if its not present or any preliminary checks


```python
class new:
  def __init__(self):
    print("You are adding the values ")
  def sum_two_values (self,a,b):
    return a+b
  def sum_n_numbers(self, *args):
    return sum(args)
```

```python
sum_obj = new()
sum_value=sum_obj.sum_two_values(2,3)
print(f"The sum values is :{sum_value}")

sum_n_numbers=new.sum_n_numbers(22,34,242,53)
print(f"The sum of all the values {sum_n_numbers}")
```
