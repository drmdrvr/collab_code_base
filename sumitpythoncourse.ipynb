```python
# we can do it using the while loop

order_values =[1122,33434,'null','dfs',1233.34,None]

sum=0


i=0


while i < len(order_values):

  if type(order_values[i])==int or type(order_values[i])==float:
    sum=sum+order_values[i]
  else:
    i=i+1 # if we miss this then it will go loop
    continue
  i=i+1

print(sum)
```

```python
#here using the whil do method:


order_values =[1122,33434,'null','dfs',1233.34,None]

sum=0


i=0


while True:
  if type(order_values[i])==int or type(order_values[i])==float:
    sum=sum+order_values[i]
  else:
    i=i+1 # if we miss this then it will go loop
    continue
  i=i+1
  if i == len(order_values)-1:
    break

print(sum)
```
